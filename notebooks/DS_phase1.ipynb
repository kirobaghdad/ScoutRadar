# Kaggle Dataset validation

```python
import os
import glob
import pandas as pd
import numpy as np

DATA_DIR = DOWNLOAD_DIR

csv_files = sorted(glob.glob(f"{DATA_DIR}/*.csv"))
tables = {}

for f in csv_files:
    name = os.path.basename(f).replace(".csv", "")
    try:
        tables[name] = pd.read_csv(f, low_memory=False)
        print(f"Loaded {name}: {tables[name].shape}")
    except Exception as e:
        print(f"Failed to load {name}: {e}")
```

# Merged Dataset Validation


## Merge Strategy

The Kaggle `games.csv` table and the API-Football `/fixtures` sample are merged at the **match level**. This follows the TA definition of merging by **adding more features about the same entity**, where the shared entity is one football match.

The merge uses the mapped competition, season, match date, and normalized home/away team names, then validates the combined table across the same eight quality dimensions.


```python
import importlib
from IPython.display import Markdown
import sys
sys.path.append('../')
from src import validation as pv

# importlib.reload(pv)  # Reload bypassed due to package restructure

merged_raw_dir = DOWNLOAD_DIR if "DOWNLOAD_DIR" in globals() else "../data/player_scores_data"
merged_api_cache_path = api_fixture_cache_path if "api_fixture_cache_path" in globals() else "../data/api_football_fixtures_sample.json"

merged_validation_results = pv.run_merged_fixture_validation(
    raw_dir=merged_raw_dir,
    api_cache_path=merged_api_cache_path,
)

merged_fixtures = merged_validation_results["merged_fixtures"]
merged_merge_summary = merged_validation_results["merge_summary"]
merged_competition_mapping = merged_validation_results["competition_mapping"]
merged_unmatched_api = merged_validation_results["unmatched_api"]
merged_unmatched_kaggle = merged_validation_results["unmatched_kaggle"]
merged_report_snippets = merged_validation_results["report_snippets"]
merged_report_lookup = merged_report_snippets.set_index("dimension")["report_text"].to_dict()

display(merged_merge_summary)
display(merged_competition_mapping)
display(merged_fixtures.head())
display(merged_unmatched_api.head(20))
display(merged_unmatched_kaggle.head(20))
```
